```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0,random_state=42)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(y)[0],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred=svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_pred,y_test)
print(acc)
```

```python
rbf=SVC(kernel='rbf')
```

```python
rbf.fit(X_train,y_train)
```

```python
y_pred1=rbf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_pred1,y_test)
print(acc)
```
